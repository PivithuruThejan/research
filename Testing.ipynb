```python
import tensorflow as tf
from tensorflow import keras
import numpy as np
import pickle
#import keras_metrics
from keras import backend as K
import numpy as np
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
```

### Data Load

```python
train_x_final = pickle.load(open('train_x_final', 'rb'))
train_y_final = pickle.load(open('train_y_final', 'rb'))

train_x_adversarial_final = pickle.load(open('train_x_adversarial_final', 'rb'))
train_y_adversarial_11_final = pickle.load(open('train_y_adversarial_11_final', 'rb'))

test_x_adversarial_final = pickle.load(open('test_x_adversarial_final', 'rb'))
test_y_adversarial_10_final = pickle.load(open('test_y_adversarial_10_final', 'rb'))
test_y_adversarial_11_final = pickle.load(open('test_y_adversarial_11_final', 'rb'))
```

```python
print(train_x_final.shape)
print(train_y_final.shape)
```

```python
train_x_final = tf.expand_dims(train_x_final, 3)

train_x_adversarial_final = tf.expand_dims(train_x_adversarial_final, 3)

test_x_adversarial_final = tf.expand_dims(test_x_adversarial_final, 3)
```

### Base Model

```python
lenet_5_model = keras.models.Sequential([
    keras.layers.Conv2D(6, kernel_size=5, strides=1,  activation='tanh', input_shape=train_x_final[0].shape, padding='same'), #C1
    keras.layers.AveragePooling2D(), #S2
    keras.layers.Conv2D(16, kernel_size=5, strides=1, activation='tanh', padding='valid'), #C3
    keras.layers.AveragePooling2D(), #S4
    keras.layers.Flatten(), #Flatten
    keras.layers.Dense(120, activation='tanh'), #C5
    keras.layers.Dense(84, activation='tanh'), #F6
    keras.layers.Dense(10, activation='softmax') #Output layer
])
```

```python
lenet_5_model.compile(optimizer='adam', loss=keras.losses.sparse_categorical_crossentropy, metrics=['accuracy'])
```

```python
lenet_5_model.fit(train_x_final, train_y_final, epochs=5)
```

```python
lenet_5_model.evaluate(test_x_adversarial_final, test_y_adversarial_10_final)
```

```python
pickle.dump(lenet_5_model, open('base_model', 'wb'))
```

```python
lenet_5_model.evaluate(test_x_adversarial_final, test_y_adversarial_10_final)
```

### Upgraded Model

```python
lenet_5_model_upgraded = keras.models.Sequential([
    keras.layers.Conv2D(6, kernel_size=5, strides=1,  activation='tanh', input_shape=train_x_adversarial_final[0].shape, padding='same'), #C1
    keras.layers.AveragePooling2D(), #S2
    keras.layers.Conv2D(16, kernel_size=5, strides=1, activation='tanh', padding='valid'), #C3
    keras.layers.AveragePooling2D(), #S4
    keras.layers.Flatten(), #Flatten
    keras.layers.Dense(120, activation='tanh'), #C5
    keras.layers.Dense(84, activation='tanh'), #F6
    keras.layers.Dense(11, activation='softmax') #Output layer
])
```

```python
lenet_5_model_upgraded.compile(optimizer='adam', loss=keras.losses.sparse_categorical_crossentropy, metrics=['accuracy'])
```

```python
lenet_5_model_upgraded.fit(train_x_adversarial_final, train_y_adversarial_11_final, epochs=5)
```

```python
lenet_5_model_upgraded.evaluate(test_x_adversarial_final, test_y_adversarial_11_final)
```

```python
pickle.dump(lenet_5_model_upgraded, open('lenet_5_model_upgraded', 'wb'))
```

### Distillation

```python
temp_y = train_y_adversarial_11_final.copy()
```

```python
temp_y.shape
```

```python
for index in range(0, len(temp_y)):
    temp_x = np.array(train_x_adversarial_final[index])
    temp_x = tf.expand_dims([temp_x], 3)
    #print(temp_x.shape)
    temp = lenet_5_model_upgraded.predict(temp_x)
    #print(temp.shape)
    temp_y[index] = np.argmax(temp)
```

```python
temp_y.shape
```

```python
lenet_5_model = keras.models.Sequential([
    keras.layers.Conv2D(6, kernel_size=5, strides=1,  activation='tanh', input_shape=train_x_adversarial_final[0].shape, padding='same'), #C1
    keras.layers.AveragePooling2D(), #S2
    keras.layers.Flatten(), #Flatten
    keras.layers.Dense(84, activation='tanh'), #F6
    keras.layers.Dense(11, activation='softmax') #Output layer
])
```

```python
lenet_5_model.compile(optimizer='adam', loss=keras.losses.sparse_categorical_crossentropy, metrics=['accuracy'])
```

```python
lenet_5_model.fit(train_x_adversarial_final, temp_y, epochs=5)
```

```python
lenet_5_model.evaluate(test_x_adversarial_final, test_y_adversarial_11_final)
```

```python
predicted_result=lenet_5_model.predict(test_x_adversarial_final)
class_preds = np.argmax(predicted_result, axis=-1)
```

```python
precission = precision_score(test_y_adversarial_11_final, class_preds,average='macro')
recall = recall_score(test_y_adversarial_11_final, class_preds,average='macro')
f1_score= 2* (precission * recall)/(precission+recall)
```

```python
print(precission)
print(recall)
print(f1_score)
```
